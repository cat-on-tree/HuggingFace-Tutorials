```python
from datasets import load_dataset

#加载数据
dataset = load_dataset(path='lansinuote/ChnSentiCorp')

dataset
```

```python
#取出训练集
data_train = dataset['train']

data_train
```

```python
#查看一个数据
data_train[0]
```

```python
#排序

#未排序的label是乱序的
print(data_train['label'][:10])

#排序之后label有序了
sorted_data = data_train.sort('label')
print(sorted_data['label'][:10])
print(sorted_data['label'][-10:])
```

```python
#洗牌

#打乱顺序
shuffled_data = sorted_data.shuffle(seed=42)

shuffled_data['label'][:10]
```

```python
#索引
data_train.select([0, 10, 20, 30, 40, 50])
```

```python
#条件筛选
def start_with_xuanze(data):
   """
   挑选出数据集中以"选择"开头的文本。
   """
   return data['text'].startswith('选择')


filtered_data = data_train.filter(start_with_xuanze)

len(filtered_data), filtered_data['text']
```

```python
#切分训练集和测试集
data_train.train_test_split(test_size=0.1)
```

```python
#把数据切分到4个桶中,均匀分配
data_train.shard(num_shards=4, index=0)
```

```python
#列的重命名
data_train.rename_column('text', 'textA')
```

```python
#删除列
data_train.remove_columns(['text'])
```

```python
#匹配

def prepend_sentence(data):
 """
 在数据集的 'text' 字段前添加 'My sentence: ' 前缀。
 """
 data['text'] = 'My sentence: ' + data['text']
 return data




modified_data = data_train.map(prepend_sentence)

modified_data['text'][:5]
```

```python
#类型转换
data_train.set_format(type='torch', columns=['label'])

data_train[0]
```

```python
#保存为csv
data_train.to_csv(path_or_buf='./data/ChnSentiCorp.csv')
```
